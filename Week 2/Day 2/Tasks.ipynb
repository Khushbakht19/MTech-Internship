### Task 1: Train a LinearRegression Model
Training a basic model using Scikit-Learn on hours-vs-marks data and making predictions for 7 and 8 hours of study.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Feature (X): Hours studied (2D array format)
X = np.array([[1], [2], [3], [4], [5]])

# Label (y): Exam scores
y = np.array([35, 50, 65, 80, 95])

# Initializing and training the model
model = LinearRegression()
model.fit(X, y)

# Making predictions for 7 and 8 hours
hours_to_predict = np.array([[7], [8]])
predictions = model.predict(hours_to_predict)

print("Predicted score for 7 hours:", round(predictions[0], 2))
print("Predicted score for 8 hours:", round(predictions[1], 2))
print("Model Slope (Marks gained per hour):", model.coef_[0])
```

### Task 2: Machine Learning Workflow Steps
Here is a short, one-sentence description for each step of the end-to-end ML workflow:

1. **Collect Data:** Gather relevant information from databases, files, or sensors to solve the problem.
2. **Clean & Preprocess:** Handle missing values, remove duplicates, and convert data into a clean format.
3. **Split Data:** Divide data into a Training set to teach the model and a Test set to evaluate it.
4. **Train Model:** Pass the training data into an algorithm so it can learn underlying patterns.
5. **Evaluate Model:** Test the model on unseen data using metrics to check its prediction accuracy.
6. **Tune Hyperparameters:** Adjust settings of the algorithm to achieve the best possible performance.
7. **Deploy Model:** Release the final trained model into a production environment for real users.
