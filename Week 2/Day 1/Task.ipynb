### Task 1: NumPy Array Operations
Creating a 1D array of 10 numbers and computing its sum, mean, maximum value, and standard deviation.

```python
import numpy as np

# Creating an array of 10 numbers
arr = np.array([12, 25, 34, 47, 51, 63, 72, 85, 90, 105])

print("Array:", arr)
print("Sum of array elements:", arr.sum())
print("Mean (Average):", arr.mean())
print("Maximum value:", arr.max())
print("Standard Deviation:", arr.std())
```

### Task 2: Loading CSV File and Inspection
Loading a dataset using Pandas to check its rows, columns, names, and initial structure.

```python
import pandas as pd

df = pd.read_csv("data.csv")

print("Dataset Shape (Rows, Columns):", df.shape)
print("\nColumn Names in the dataset:\n", df.columns.tolist())
print("\nFirst 5 Rows of the data:")
print(df.head())
```

### Task 3: Filtering Data
Filtering rows based on a specific condition (e.g., viewing records where Salary is greater than 50,000).

```python
# Filtering rows where Salary is more than 50000
# (Note: Change 'Salary' to your actual numeric column name if your CSV is different)
filtered_df = df[df["Salary"] > 50000]
print(filtered_df)
```

### Task 4: Handling Missing Data
Finding missing values (NaN) in a column and replacing them with the column mean.

```python
# Checking total missing values in the column
# (Note: Change 'Bonus' to your target column name containing missing values)
print("Missing values before cleaning:", df["Bonus"].isnull().sum())

# Filling missing entries with the mean value of that column
mean_value = df["Bonus"].mean()
df["Bonus"] = df["Bonus"].fillna(mean_value)

print("Missing values after cleaning:", df["Bonus"].isnull().sum())
print("\nCleaned Column:\n", df["Bonus"].head())
```

### Task 5: Grouping Data
Grouping data by a categorical column (like 'City') and finding the mean of a numeric column.

```python
# Grouping data by City and calculating average Salary
# (Note: Change 'City' and 'Salary' based on your CSV columns)
summary = df.groupby("City")["Salary"].mean()
print(summary)
```
